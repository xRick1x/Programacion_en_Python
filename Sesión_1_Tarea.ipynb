3.1 驴Cu谩l es la tecnolog铆a m谩s usada en Per煤, seg煤n la base de datos?

```python
import pandas as pd

# Calcular la suma total de cada tecnolog铆a en todo el DataFrame
total_2g = df['2G'].sum()
total_3g = df['3G'].sum()
total_4g = df['4G'].sum()
total_5g = df['5G'].sum()

# Encontrar la tecnolog铆a m谩s utilizada
tecnologia_mas_utilizada = max([('2G', total_2g), ('3G', total_3g), ('4G', total_4g), ('5G', total_5g)], key=lambda x: x[1])[0]

print("La tecnolog铆a m谩s utilizada en Per煤 es:", tecnologia_mas_utilizada)
```

3.2 驴Cu谩l es la tecnolog铆a con menor porcentaje de uso? 驴Cu谩l es su porcentaje de uso?

```python
import pandas as pd

# Calcular el total de todas las tecnolog铆as
total_tecnologias = df[['2G', '3G', '4G', '5G']].sum().sum()

# Calcular el porcentaje de uso de cada tecnolog铆a
porcentaje_uso = df[['2G', '3G', '4G', '5G']].sum() / total_tecnologias  * 100

# Encontrar la tecnolog铆a con el menor porcentaje de uso
tecnologia_menor_porcentaje = porcentaje_uso.idxmin()
menor_porcentaje = porcentaje_uso.min()

print("La tecnolog铆a con el menor porcentaje de uso es:", tecnologia_menor_porcentaje)
print("El porcentaje de uso m谩s bajo es:", menor_porcentaje, "%")
```
