```python
import pandas as pd
import numpy as np


data_total['FECHA DE REGISTRO'] = pd.to_datetime(data_total['FECHA DE REGISTRO'])
# Se ordenar el DataFrame por fecha
data_total_sorted = data_total.sort_values(by='FECHA DE REGISTRO')

# Inicializar la variable para almacenar el último valor válido (menor o igual a 100)
last_valid_value = None

# Función para reemplazar valores mayores a 100 con el último valor válido
def replace_greater_than_100(row):
    global last_valid_value
    if row['PRECIO DE VENTA (SOLES)'] > 100:
        return last_valid_value
    else:
        last_valid_value = row['PRECIO DE VENTA (SOLES)']
        return row['PRECIO DE VENTA (SOLES)']

# Almacenando los nuevos datos en Dataframe
data_total_modificado = data_total_sorted.copy()

# Aplicar la función a cada fila del nuevo DataFrame
data_total_modificado['PRECIO DE VENTA (SOLES)'] = data_total_modificado.apply(replace_greater_than_100, axis=1)

print(data_total_modificado)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data_filtrada_2['FECHA DE REGISTRO'] = pd.to_datetime(data_filtrada_2['FECHA DE REGISTRO'])

# Ordenar el DataFrame por 'fecha'
data_filtrada_2 = data_filtrada_2.sort_values(by='FECHA DE REGISTRO')

# Agrupar por 'distrito', 'producto', y 'fecha' y calcular el precio promedio diario
precio_promedio_diario = data_filtrada_2.groupby(['DISTRITO', 'PRODUCTO', 'FECHA DE REGISTRO'])['PRECIO DE VENTA (SOLES)'].mean().reset_index()

# Crear el gráfico de líneas para cada distrito y producto
distritos = precio_promedio_diario['DISTRITO'].unique()
productos = ['GASOHOL REGULAR', 'GASOHOL PREMIUM', 'GASOLINA 98']

fig, ax = plt.subplots(figsize=(14, 8))

for distrito in distritos:
    for producto in productos:
        data_distrito_producto = precio_promedio_diario[(precio_promedio_diario['DISTRITO'] == distrito) & (precio_promedio_diario['PRODUCTO'] == producto)]
        ax.plot(data_distrito_producto['FECHA DE REGISTRO'], data_distrito_producto['PRECIO DE VENTA (SOLES)'], label=f'{producto} en {distrito}')

ax.set_title('Tendencia del Precio de Productos por Distrito en Lima')
ax.set_xlabel('FECHA DE REGISTRO')
ax.set_ylabel('PRECIO DE VENTA (SOLES)')
ax.legend(title='PRODUCTO y DISTRITO', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

data_filtrada_2['FECHA DE REGISTRO'] = pd.to_datetime(data_filtrada_2['FECHA DE REGISTRO'])

# Filtrar el DataFrame para conservar solo el distrito de San Borja
data_san_borja = data_filtrada_2[data_filtrada_2['DISTRITO'] == 'SAN BORJA']

# Ordenar el DataFrame por 'fecha'
data_san_borja = data_san_borja.sort_values(by='FECHA DE REGISTRO')

# Agrupar por 'producto' y 'fecha' y calcular el precio promedio diario
precio_promedio_diario_san_borja = data_san_borja.groupby(['PRODUCTO', 'FECHA DE REGISTRO'])['PRECIO DE VENTA (SOLES)'].mean().reset_index()

# Crear el gráfico de líneas para cada producto en el distrito de San Borja
productos = ['GASOHOL REGULAR', 'GASOHOL PREMIUM', 'GASOLINA 98']

fig, ax = plt.subplots(figsize=(14, 8))

for producto in productos:
    data_producto = precio_promedio_diario_san_borja[precio_promedio_diario_san_borja['PRODUCTO'] == producto]
    ax.plot(data_producto['FECHA DE REGISTRO'], data_producto['PRECIO DE VENTA (SOLES)'], label=producto)

ax.set_title('Tendencia del Precio de Productos en San Borja')
ax.set_xlabel('FECHA DE REGISTRO')
ax.set_ylabel('PRECIO DE VENTA (SOLES)')
ax.legend(title='PRODUCTO')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
data_filtrada_2['FECHA DE REGISTRO'] = pd.to_datetime(data_filtrada_2['FECHA DE REGISTRO'])

# Filtrar el DataFrame para conservar solo el distrito de San Borja
data_san_borja = data_filtrada_2[data_filtrada_2['DISTRITO'] == 'SAN BORJA']
data_san_borja = data_san_borja.sort_values(by='FECHA DE REGISTRO')

# Agrupar por 'producto', 'fecha', y 'vendedor' y calcular el precio promedio diario
precio_promedio_diario_san_borja = data_san_borja.groupby(['PRODUCTO', 'FECHA DE REGISTRO', 'RAZÓN SOCIAL'])['PRECIO DE VENTA (SOLES)'].mean().reset_index()

# Identificar los cuatro vendedores más baratos para cada producto y fecha
def obtener_vendedores_mas_baratos(grupo):
    return grupo.nsmallest(4, 'PRECIO DE VENTA (SOLES)')

# Obtener los cuatro vendedores más baratos en cada grupo de 'producto' y 'fecha'
vendedores_mas_baratos = precio_promedio_diario_san_borja.groupby(['PRODUCTO', 'FECHA DE REGISTRO']).apply(obtener_vendedores_mas_baratos).reset_index(drop=True)

# Crear el gráfico de líneas para cada producto y los cuatro vendedores más baratos en el distrito de San Borja
productos = ['GASOHOL REGULAR', 'GASOHOL PREMIUM', 'GASOLINA 98']

fig, ax = plt.subplots(figsize=(14, 8))

for producto in productos:
    data_producto = vendedores_mas_baratos[vendedores_mas_baratos['PRODUCTO'] == producto]
    for vendedor in data_producto['RAZÓN SOCIAL'].unique():
        data_vendedor = data_producto[data_producto['RAZÓN SOCIAL'] == vendedor]
        ax.plot(data_vendedor['FECHA DE REGISTRO'], data_vendedor['PRECIO DE VENTA (SOLES)'], label=f'{producto} - {vendedor}')

ax.set_title('Tendencia del Precio de los 4 Vendedores Más Baratos en San Borja')
ax.set_xlabel('FECHA DE REGISTRO')
ax.set_ylabel('PRECIO DE VENTA (SOLES)')
ax.legend(title='PRODUCTO y Vendedor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
